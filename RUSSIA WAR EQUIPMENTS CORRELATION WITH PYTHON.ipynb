```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration of the plot we will create

#Read in the data
df=pd.read_csv("russia_losses_equipment.csv") ####
```

```python
#to visualize our correlation matrix
#it's nice to visualize this because it is tough reading from the correlation matrix
correlation_matrix=df.corr(method="pearson")
sns.heatmap(correlation_matrix,annot=True)
plt.suptitle("Correlation Matrix")
plt.ylabel("Russia losses equipment")
plt.xlabel("Russia losses equipment")
plt.show()
```

```python
df_numerized=df 
for i in df_numerized.columns:
    if (df_numerized[i].dtype=="object"): 
        df_numerized[i]=df_numerized[i].astype('category')
        df_numerized[i]=df_numerized[i].cat.codes
df_numerized
```

```python
df
```

```python
#to view our correlation matrix 
df_numerized.corr() 
```

```python
#to visualize our correlation matrix
correlation_matrix=df_numerized.corr(method="pearson")
sns.heatmap(correlation_matrix,annot=True)
plt.suptitle("Correlation Matrix")
plt.ylabel("Russia losses equipment")
plt.xlabel("Russia losses equipment")
plt.show()
```

```python
#Organize to see the one that has the highest correlation
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs>0.5)]
high_corr
```
